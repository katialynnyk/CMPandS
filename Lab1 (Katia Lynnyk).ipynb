```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
img_bw = cv2.imread("./cat.jpg", cv2.IMREAD_GRAYSCALE)
print(img_bw)
```

```
cv2.imwrite("./cat_write.png", img_bw)
```

```
img_rgb = cv2.imread("./rgb.jpg")
cv2.imwrite("./rgb_write.png", img_rgb)
```

```
plt.figure(figsize=(10, 10))
plt.imshow(img_bw, cmap='gray')
plt.show();
```

```
rgb = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 10))
plt.imshow(rgb)
plt.show();
```

```
rgb_hsv = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2HSV)
plt.figure(figsize=(10, 10))
plt.imshow(rgb_hsv)
plt.show();
```

```
plt.figure(figsize=(10, 10))

plt.subplot(131);
R = np.zeros(rgb.shape, dtype='uint8')
R[:, :, 0] = rgb[:, :, 0]
plt.imshow(R)

plt.subplot(132);
G = np.zeros(rgb.shape, dtype='uint8')
G[:, :, 1] = rgb[:, :, 1]
plt.imshow(G)

plt.subplot(133);
B = np.zeros(rgb.shape, dtype='uint8')
B[:, :, 2] = rgb[:, :, 2]
plt.imshow(B)

plt.show()
```

```
plt.figure(figsize=(10, 10))

plt.subplot(131);
H = np.zeros(rgb_hsv.shape, dtype='uint8')
H[:, :, 0] = rgb_hsv[:, :, 0]
plt.imshow(H)

plt.subplot(132);
S = np.zeros(rgb_hsv.shape, dtype='uint8')
S[:, :, 1] = rgb_hsv[:, :, 1]
plt.imshow(S)

plt.subplot(133);
V = np.zeros(rgb_hsv.shape, dtype='uint8')
V[:, :, 2] = rgb_hsv[:, :, 2]
plt.imshow(V)

plt.show()
```

```
crop = img_bw.shape
x = crop[0]/2
y = crop[1]/2

img_crop_bw = img_bw[0 : int(x), 0 : int(y)]
plt.imshow(img_crop_bw, cmap = "gray")
plt.show()
```

```
cv2.imwrite("./img_crop_bw.png", img_crop_bw)
```
